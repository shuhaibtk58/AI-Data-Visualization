```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('studentdata.csv')
data
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(x="Name", y="Marks", data=data)
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x="Name", y="Attendence", data=data)
plt.title("Bar Plot")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Class", y="Marks", data=data, hue="Age")
plt.title("Scatter Plot")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="Class", y="Marks", data=data)
plt.title("Box Plot")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.pointplot(x="Age", y="Marks", data=data, hue="Gender")
plt.title("Point Plot")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x="Grade", y="Marks", data=data, hue="Gender")
plt.title("Violin Plot")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data["Age"], kde=True)
plt.title("Distribution Plot")
plt.show()
```
